## Clean Census tract

```python
def censusTract(x):
    """Cleans 'Geography' column into Census tract"""
    
    x_split = x.split(',')
    x_split = x_split[0].split(' ')
    return x_split[2]
```

```python
a = 'Census Tract 19, Bronx County, New York'
a.split(',')[1].split(' ')
```

```python
def boroughCensusTract(x):
    """Cleans 'Geography' column into Borough/Census tract of form (BoroughName)_(CensusTract)
    For example, Census Tract 19, Bronx County, New York ---> Bronx_19"""
    
    x_split = x.split(',')
    tract_split = x_split[0].split(' ')
    county_split = x_split[1].split(' ')
    
    #Manhattan
    if (county_split[1] == 'New'):
        return 'Manhattan_' + tract_split[2]
    #Brooklyn
    elif (county_split[1] == 'Kings'):
        return 'Brooklyn_' + tract_split[2]
    #Queens
    elif (county_split[1] == 'Queens'):
        return 'Queens_' + tract_split[2]
    #Bronx
    elif (county_split[1] == 'Bronx'):
        return 'Bronx_' + tract_split[2]
    #Staten Island
    elif (county_split[1] == 'Richmond'):
        return 'StatenIsland_' + tract_split[2]
    #find my mistakes
    else:
        return 'Check'
```

```python
df['borough_census_tract'] = df.Geography.apply(lambda x: boroughCensusTract(x))
```

# Remove columns that we won't need

```python
df_clean = df[['borough_census_tract', 'Total; Estimate; Total population',
              'Total; Estimate; Median age (years)',
               'Total; Margin of Error; EDUCATIONAL ATTAINMENT - Population 25 years and over',
               'Total; Estimate; EDUCATIONAL ATTAINMENT - Population 25 years and over - Bachelor\'s degree',
               'Total; Estimate; Median income (dollars)'
              ]]
```

```python
#Rename columns
df_clean.columns = ['census_tract', 'total_population', 'median_age',
                    'total_population_25yrsandover','total_bachelors_degree','median_income'
                   ]
```
