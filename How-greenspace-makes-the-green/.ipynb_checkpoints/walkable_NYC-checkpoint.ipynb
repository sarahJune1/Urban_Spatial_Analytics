```python
#Read in walkability map
NY_df = gpd.read_file("NY_shape/NY.shp")
NY_df.head()
```

```python
#Read in County fips data
fips_codes = pd.read_csv("https://www2.census.gov/geo/docs/reference/codes/files/national_county.txt", 
                         names = ["State", "SFIPS", "CFIPS", "County", "FIPS Status"], 
                         dtype={'SFIPS': str, 'CFIPS': str})
fips_codes['FIPS'] = fips_codes['SFIPS'] + fips_codes['CFIPS']
```

```python
#Read in street map
NYC_streets = gpd.read_file("NYC_streets/geo_export_11a987d5-4699-4ae8-9876-b4c7945d3431.shp")
NYC_streets.head()
```

```python
# Full NY Shapefile
fig, ax = plt.subplots(figsize = (14,12))
ax = NY_df.plot(column = 'NatWalkInd', ax = ax, cmap = 'viridis', legend = True)
plt.title("Walkability Index")
plt.show()
```

```python
#Create FIPS code
NY_df['FIPS'] = NY_df['SFIPS'] + NY_df['CFIPS']

#Merge data with fips_codes so we have names as well
NYC_df = NY_df.merge(fips_codes, how = 'inner', left_on= 'FIPS', right_on = 'FIPS')
NYC_df = NYC_df.drop(['CFIPS_y', 'SFIPS_y', 'CBSA_Name'], axis = 1)
NYC_df = NYC_df.rename(columns = {'CFIPS_x': 'CFIPS', 'SFIPS_x': 'SFIPS'})
NYC_df['County'] = NYC_df['County'].apply(lambda x: x[:-7])
NYC_df.head()
```

```python
#NYC Counties: Kings, New York, Bronx, Queens, Richmond
NYC_df = NYC_df.loc[NYC_df['County'].isin(['Kings', 'New York', 'Bronx', 'Queens', 'Richmond'])]
NYC_df = NYC_df.loc[NYC_df['TOTPOP10'] != 0]
NYC_df.head()
```

```python
fig, ax = plt.subplots(figsize = (14,12))
NYC_df.plot(column = 'NatWalkInd', ax = ax, cmap = 'viridis', legend = True)
plt.show()
```
