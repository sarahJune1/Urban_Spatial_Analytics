```python
import geopandas as gpd
import pylab as plt
import pandas as pd
import numpy as np
import os
import io
import osmnx as ox
import networkx as nx
from sklearn.neighbors import KDTree
%matplotlib inline
```

```python
NYC = ox.graph_from_address('East Village, New York', network_type='walk')
```

```python
ox.plot_graph(NYC)
```

```python
nodes, _ = ox.graph_to_gdfs(NYC)
nodes.head()
```

```python
# Choose two points on map to find shortest distance
stern = ox.geocode("44 West 4th Street, New York")
theater = ox.geocode("425 Lafayette St, New York")
```

```python
#Tree to traverse and findest nearest node to locations. (The blue dots on map above)
tree = KDTree(nodes[['y', 'x']], metric='euclidean')

stern_idx = tree.query([stern], k=1, return_distance=False)[0]
theater_idx = tree.query([theater], k=1, return_distance=False)[0]

closest_stern = nodes.iloc[stern_idx].index.values[0]
closest_theater = nodes.iloc[theater_idx].index.values[0]
```

```python
fig, ax = ox.plot_graph(NYC, fig_height=10, fig_width=10, show=False, close=False,
                       edge_color='grey')
ax.scatter(stern[1], stern[0], c = 'red', s= 100)
ax.scatter(theater[1], theater[0], c = 'blue', s = 100)
ax.scatter(NYC.node[closest_stern]['x'],
          NYC.node[closest_stern]['y'],
          c = 'green', s = 100)
ax.scatter(NYC.node[closest_theater]['x'], NYC.node[closest_theater]['y'],
          c = 'green', s = 100)
plt.show()
```

```python
route = nx.shortest_path(NYC, closest_stern, closest_theater)

fig, ax = ox.plot_graph_route(NYC, route, fig_height=10, fig_width=10, show=False, close=False,
                             edge_color = 'grey', orig_dest_node_color='green',
                              route_color='red')

ax.scatter(stern[1], stern[0], c = 'red', s= 100)
ax.scatter(theater[1], theater[0], c = 'blue', s = 100)

plt.show()
```
